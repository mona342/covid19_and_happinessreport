#  Aggregating the rows by the country

```python
df.aggregated=df.groupby("Country/Region").sum()
```

#  caculating the first derivative of the curve

```python
df.aggregated.loc["China"].diff().plot()
```

```python
df.aggregated.loc["China"].diff().max()
#In only 24 hrs, the difference was 15136
```

```python
df.aggregated.loc["Italy"].diff().max()
#In only 24 hrs, the difference was 6557
```

```python
df.aggregated.loc["Spain"].diff().max()
#In only 24 hrs, the difference was 9630
```

```python
countries = list(df.aggregated.index)
max_infection_rates = []
for c in countries :
    max_infection_rates.append(df.aggregated.loc[c].diff().max())
max_infection_rates
```

```python
df.aggregated["max_infection_rates"] = max_infection_rates
```

```python
df.aggregated.head()
```

```python
corona_data = pd.DataFrame(df.aggregated["max_infection_rates"])
```

```python
corona_data.head()
```
